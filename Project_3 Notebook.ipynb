## Task 1: Data Exploration

Checking Null Values , release_date into datatime , best & worst total gross

```python
sns.heatmap(data.isnull(),yticklabels=False)
```

```python
def year(x):
    return x.split("-")[0]
```

```python
data['year']=data['release_date'].apply(year).astype(int)
```

```python
data
```

```python
data.nlargest(n=1,columns=['total_gross'])
```

```python
data.nsmallest(n=1,columns=['total_gross'])
```

## Task 2: Visualize Year Distribution Plot

```python
fig,ax=plt.subplots(figsize=(10,8))
sns.distplot(data['year'])
```

```python
data['year'].value_counts().plot(kind='pie', fontsize=10, autopct='%1.1f%%', 
                                               figsize=(20,10), shadow=True, startangle=90, cmap='Blues')
```

```python
data.groupby('year')['year'].count().plot(kind='bar',figsize=(20,18),color='blue')
```

## Task 3: Does Movies Income Decreased or Increased?

```python
fig,ax=plt.subplots(figsize=(10,8))
sns.scatterplot(data.year,data.total_gross)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,8))
sns.lineplot(x='year',y='total_gross',data=data)
```

```python
fig=sns.factorplot(x='year',y='total_gross',size=6,aspect=2,kind='bar',data=data)
plt.xticks(rotation=50)
```
